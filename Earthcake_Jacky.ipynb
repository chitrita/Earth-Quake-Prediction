# Load dataset

Due to the high number of rows, it is impossible for now to download all. We will limit the number of rows during the exploration phase.

```
df = pd.read_csv('train.csv', names=['acoustic','time'], skiprows=15000000, nrows=40000000)
```

```
df.head()
```

# Vizualisation

```
fig, ax1 = plt.subplots()
ax1.set_title('Signal', size=18)
ax1.set_xlabel('lines')
ax1.set_ylabel('acoustic')
ax1.plot(range(len(df.acoustic)), df.acoustic.values, color = 'red')
ax1.tick_params(axis='y', size=15)

# instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()

ax2.set_ylabel('time before earthquake')  # we already handled the x-label with ax1
ax2.plot(range(len(df.acoustic)), df.time.values)
ax2.tick_params(axis='y', size=15)
fig.set_figwidth(18)
fig.set_figheight(10)
```
